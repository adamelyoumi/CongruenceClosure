```python
EXAMPLE = "f(a) = b & f(f(a)) = c & g(c, f(b)) != g(c, c)"
```

```python
inp = "f(a) =b & f(b)= f(g(a)) &f(a)!=b"
inp2 = "f(a, b) =b & f(b,c)= f(g(a), b) &f(a,c)!=b"
```

```python
trimmed = inp.replace(" ", "")
clauses = trimmed.split("&")
clauses
```

```python
clauses_dict = {i: [] for i in range(-1,len(clauses)-1)}
for i in range(len(clauses)):
    c = clauses[i]
    if "!=" in c:
        left_ineq, right_ineq = c.split("!=")
        clauses_dict[-1] = [left_ineq, right_ineq]
    else:
        clauses_dict[i] = c.split("=")
```

```python
clauses_dict
```

```python
left_ineq, right_ineq
```

```python
terms = []
functions, literals = [], []
for k,v in clauses_dict.items():
    terms += v

terms
```

```python
def extract_function_names(expression):
    pattern = r'(\w+)\('
    matches = re.findall(pattern, expression)
    return matches
```

```python
def extract_literals(expression):
    pattern = r'\b\w+\b(?=[,)])'
    matches = re.findall(pattern, expression)
    return matches
```

```python
def apply(s, func):
    return func + "(" + s + ")"
```

```python
for t in terms:
    functions += extract_function_names(t)
    literals += extract_literals(t)
    
functions = list(set(functions))
literals = list(set(literals))
functions, literals
```

```python
def function_chain(expression):
    result = []
    while True:
        m = re.search(r'(\w+)\((.+)\)', expression)
        if m is None:
            break
        result.append(expression)
        expression = m.group(2)
    result.append(expression)
    return result

# Test case
print(function_chain("f(f(f(f(g(x)))))"))
```

```python
T = []
for t in terms:
    T += function_chain(t)
    
T = (list(set(T)))
eq_classes = [[k] for k in T]
print(eq_classes)
```

```python
def CC(clauses_dict: dict, eq_classes):
    left_ineq, right_ineq = clauses_dict[-1]
    clauses_dict.pop(-1)
    for k, v in clauses_dict.items():
        left, right = v
        left_c, right_c = -1, -1
        for i, c in enumerate(eq_classes):
            if left in c:
                left_c = i
            if right in c:
                right_c = i
                
        if left_c != right_c:
            eq_classes[left_c] = eq_classes[left_c] + eq_classes[right_c]
            eq_classes.pop(right_c)
            
        for i, c in enumerate(eq_classes):
            if left_ineq in c and right_ineq in c:
                return "UNSAT"
            
    return("SAT")
    
CC(clauses_dict, eq_classes)
```
